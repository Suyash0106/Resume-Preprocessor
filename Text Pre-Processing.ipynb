## 1. Introduction

This program reads data from a text document which contains resumes of about 30,000 candidates. This data is extracted and structured into XML and JSON format. The documents containing the formatted output are stored in "29279208_output.xml" and "29279208_output.json" files respectively.

As different identifiers are used to denote the same key, regular expressions for finding all the  possible identifiers is created. A dictionary containing different identifiers and their count is created. This dictionary is sorted according to the count of identifiers. This dictionary is then used to identify the potential identifiers that refer to the same key. 

Once all the potential identifiers are identified, regular expression for the identifiers is created and these identifiers are replaced by a single unique identifier that will refer to its key. A new preprocessed file "29279208_preprocessedData.txt" is created and unique identifier of keys and its values are stored in it.

This file is then used for extracting values corresponding to the keys and feeding these values into XML and JSON documents.

The following code explains the step by step process to convert the text document to XML and JSON documents.

## 3.  Extracting the data and creating pre-processed file

### 3.a Load data from input file

Load data from the file "29279208.dat".<br>
Create a list containing individual applications by splitting the data. The list "data" contains all the individual resumes.

```python
file = open("29279208.dat","r")
data = file.read()
data = data.split("------------------------------")
data.pop(-1)
file.close()
print("OK")
```

### 3.b Create dictionary of  arguments

The below regular expression is used for the identifiers. It finds all the words that are on the newline and ends with ":" or "/".

**"(^([a-z]*_?[a-z]+|[A-Z_]+)[^\n]?[A-Za-z]*(:|/))"**

A dictionary "setx" is created that contains all the identifiers and their count. This dictionary is sorted according to the count of identifiers. Now, this sorted output is analysed for creating the regular expressions for the potential identifiers of the key.

```python
set_keys= dict()
for each in data:
    keys= re.findall(r"(^([a-z]*_?[a-z]+|[A-Z_]+)[^\n]?[A-Za-z]*(:|/))", each, re.MULTILINE)
    
    for item in keys:
        if set_keys.get(item[0]):
            set_keys[item[0]] +=1
        else:
            set_keys[item[0]] = 1

sorted_set_keys = sorted(set_keys.items(), key = lambda a: a[1], reverse=True) 

for each in sorted_set_keys:
    print(each[0]+" "+str(each[1]))
```

### 3.c Create a preprocessed file

Using the dictionary created in **3.b**, create regular expressions for potential identifier of the keys. Once the potential identifiers are found, substitute them with unique identifier for the particular key and store this info in the file "29279208_preprocessedData.txt". This pre-processed file is then used for extracting values of the keys.

**Note:**
1. This file is created in an "append+" mode and gets created in the same folder in which this jupyter file is present.
2. **"OK"** is printed on the console to signal the file is successfully generated.
3. Expected execution time is 2-5 minutes.
4. **"%%END%%"** is a unique identifier used in this file to mark the end of each resume.

```python
newdata = open("29279208_preprocessedData.txt","a+")

for each in data:
    aboutCompany = re.sub(r"^(ABOUT\sCOMPANY|about_company|COMPANYS_INFO|_info|ABOUT)[:/]?","%%ABOUT_COMPANY%%:", each,flags=re.MULTILINE)
    deadline = re.sub(r"^(APPLICATION_D.*|DEAD.*|deadline):","%%DEADLINE%%:", aboutCompany, flags=re.MULTILINE)
    jobProcedure = re.sub(r"^(JOB_PROCS?|procedures?|PROCEDURES?):","%%JOB_PROCEDURE%%:", deadline, flags=re.MULTILINE)
    jobDescription = re.sub(r"^(JOB[\s_]DESC.*|job_desc|_description|DESCRIPTION):","%%JOB_DESCRIPTION%%:", jobProcedure, flags=re.MULTILINE)
    location = re.sub(r"^(_?LOC.*?|JOB_LOC):","%%LOCATION%%:", jobDescription, flags=re.MULTILINE)
    responsibilities = re.sub(r"^(JOB[\s_]?RESP.*|RESP.*|responsibilities):","%%JOB_RESPONSIBILITIES%%:", location, flags=re.MULTILINE)
    qualifications = re.sub(r"^(REQ.*?QUAL.*|QUAL.*|qualifications):","%%QUALIFICATIONS%%:", responsibilities, flags=re.MULTILINE)
    salary = re.sub(r"^(SALARY|REMUNERATION|JOB_SAL|salary|remuneration)[:/]","%%SALARY%%:", qualifications, flags=re.MULTILINE)
    title = re.sub(r"^(JOB[\s_]?T.*|_TTL|title|TITLES?):","%%JOB_TITLE%%:", salary, flags=re.MULTILINE)
    startDate = re.sub(r"^(START[\s_]?DA.*|DATE_START|start_date|DATES)[:/]","%%START_DATE%%:", title, flags=re.MULTILINE)
    jobId = re.sub(r"^(ID):","%%ID%%:", startDate, flags=re.MULTILINE)    
    amp = re.sub(r"&", "&amp;",jobId,flags=re.MULTILINE)
    gt = re.sub(r">", "&gt;",amp,flags=re.MULTILINE)
    lt = re.sub(r"<", "&lt;",gt,flags=re.MULTILINE)
    newdata.write(lt)
    newdata.write("%%END%%")
    newdata.write("\n------------------------------")

newdata.close()
print("OK")
```

### 4.a Load data from pre-processed file

Read the pre-processed file "29279208_preprocessedData.txt" and store the individual resumes in the list "data".

### 4.b Write the data in XML and JSON files

1. Open the "29279208_output.xml" and "29279208_output.json" files for writing XML and JSON data respectively. Also, create "listingJSON" and "listings" dictionaries to fit the json values appropriately.

```python
# Open the "29279208_output.xml" file for writing the XML output
output_xml = open("29279208_output.xml","w+")

# Open the "29279208_output.json" file for writing the JSON output
output_json = open("29279208_output.json","w+")

# This dictionary is created according to the sample format which contains the dictionary of listings
listingJSON = dict()

# This dictionary contains the list of resumes "listing".
listings = dict()

# This list contains resumes.
listing=[]

print("OK")
```

Each resume is read from the list "data". For each resume, regular expressions are used for extracting the values of the following keys:

* id
* location
* jobDescription
* jobResponsibility
* jobTitle
* aboutCompany
* deadLine
* jobProcedure
* jobQualifications
* salary
* startDate

Each regular expression follows similar pattern. For example:

location =** ^(%%LOCATION%%:)((.*\s)*?)(%%ABOUT_COMPANY%%:|%%DEADLINE%%:|%%JOB_PROCEDURE%%:|%%JOB_DESCRIPTION%%:|%%JOB_RESPONSIBILITIES%%:|%%QUALIFICATIONS%%:|%%SALARY%%:|%%JOB_TITLE%%:|%%START_DATE%%:|OPEN TO/|%%END%%)**

This regex is used to search value of location key. This regex is mainly divided into 3 groups:
1. ^(%%LOCATION%%:)
2. ((.*\s)*?)
3. (%%ABOUT_COMPANY%%:|%%DEADLINE%%:|%%JOB_PROCEDURE%%:|%%JOB_DESCRIPTION%%:|%%JOB_RESPONSIBILITIES%  %:|%%QUALIFICATIONS%%:|%%SALARY%%:|%%JOB_TITLE%%:|%%START_DATE%%:|OPEN TO/|%%END%%)

Group 1 is used to identify the key in the pre-processed file. Group 2  is used to extract the value that is present between 2 keys. Group 3 is used to identify the next key after "%%LOCATION%%" so that the value that lies between the "%%LOCATION%%" and the next key can be extracted.

This pattern of regex for extracting values of keys is similar for all the keys.

Values of keys for each resume is extracted and fitted appripriately into XML and JSON files respectively. The version of XML used in "1.0" and encoding used is "UTF-8". 

For JSON, the "json" library is imported and **"json.dumps()"** function is used to create JSON object from python dictionaries.

**Note:**
1. **"OK"** is printed on the console to indicate successful creation of files.
2. Both the XML and JSON files are created simultaneously.
3. Expected execution time is 2-5 minutes.

```python
# XML file creation
output_xml.write("<?xml version=\"1.0\" encoding=\"UTF-8\" ?>\n")
output_xml.write("<listings>\n")

for each in data:
    id = re.search(r'%%ID%%: (\d+)', each,re.MULTILINE).group(1)
    location = re.search(r'^(%%LOCATION%%:)((.*\s)*?)(%%ABOUT_COMPANY%%:|%%DEADLINE%%:|%%JOB_PROCEDURE%%:|%%JOB_DESCRIPTION%%:|%%JOB_RESPONSIBILITIES%%:|%%QUALIFICATIONS%%:|%%SALARY%%:|%%JOB_TITLE%%:|%%START_DATE%%:|OPEN TO/|%%END%%)', each,re.MULTILINE)
    jobDescription = re.search(r'^(%%JOB_DESCRIPTION%%:)((.*\s)*?)(%%LOCATION%%:|%%ABOUT_COMPANY%%:|%%DEADLINE%%:|%%JOB_PROCEDURE%%:|%%JOB_RESPONSIBILITIES%%:|%%QUALIFICATIONS%%:|%%SALARY%%:|%%JOB_TITLE%%:|%%START_DATE%%:|OPEN TO/|%%END%%)', each,re.MULTILINE)
    jobResponsibility = re.search(r'^(%%JOB_RESPONSIBILITIES%%:)((.*\s)*?)(%%JOB_TITLE%%:|%%LOCATION%%:|%%ABOUT_COMPANY%%:|%%DEADLINE%%:|%%JOB_PROCEDURE%%:|%%JOB_DESCRIPTION%%:|%%QUALIFICATIONS%%:|%%SALARY%%:|%%START_DATE%%:|OPEN TO/|%%END%%)', each,re.MULTILINE)
    jobTitle = re.search(r'^(%%JOB_TITLE%%:)((.*\s)*?)(%%ABOUT_COMPANY%%:|%%DEADLINE%%:|%%JOB_PROCEDURE%%:|%%JOB_DESCRIPTION%%:|%%LOCATION%%:|%%JOB_RESPONSIBILITIES%%:|%%QUALIFICATIONS%%:|%%SALARY%%:|%%START_DATE%%:|OPEN TO/|%%END%%)', each,re.MULTILINE)
    aboutCompany = re.search(r'^(%%ABOUT_COMPANY%%:)((.*\s)*?)(%%JOB_PROCEDURE%%:|%%DEADLINE%%:|%%JOB_DESCRIPTION%%:|%%LOCATION%%:|%%JOB_RESPONSIBILITIES%%:|%%QUALIFICATIONS%%:|%%SALARY%%:|%%JOB_TITLE%%:|%%START_DATE%%:|OPEN TO/|%%END%%)', each,re.MULTILINE)
    deadLine = re.search(r'^(%%DEADLINE%%:)((.*\s)*?)(%%ABOUT_COMPANY%%:|%%JOB_PROCEDURE%%:|%%JOB_DESCRIPTION%%:|%%LOCATION%%:|%%JOB_RESPONSIBILITIES%%:|%%QUALIFICATIONS%%:|%%SALARY%%:|%%JOB_TITLE%%:|%%START_DATE%%:|OPEN TO/|%%END%%)', each,re.MULTILINE)
    jobProcedure = re.search(r'^(%%JOB_PROCEDURE%%:)((.*\s)*?)(%%ABOUT_COMPANY%%:|%%DEADLINE%%:|%%JOB_DESCRIPTION%%:|%%LOCATION%%:|%%JOB_RESPONSIBILITIES%%:|%%QUALIFICATIONS%%:|%%SALARY%%:|%%JOB_TITLE%%:|%%START_DATE%%:|OPEN TO/|%%END%%)', each,re.MULTILINE)
    jobQualifications = re.search(r'^(%%QUALIFICATIONS%%:)((.*\s)*?)(%%JOB_PROCEDURE%%:|%%ABOUT_COMPANY%%:|%%DEADLINE%%:|%%JOB_DESCRIPTION%%:|%%LOCATION%%:|%%JOB_RESPONSIBILITIES%%:|%%SALARY%%:|%%JOB_TITLE%%:|%%START_DATE%%:|OPEN TO/|%%END%%)', each,re.MULTILINE)
    salary = re.search(r'^(%%SALARY%%:)((.*\s)*?)(%%QUALIFICATIONS%%:|%%JOB_PROCEDURE%%:|%%ABOUT_COMPANY%%:|%%DEADLINE%%:|%%JOB_DESCRIPTION%%:|%%LOCATION%%:|%%JOB_RESPONSIBILITIES%%:|%%JOB_TITLE%%:|%%START_DATE%%:|OPEN TO/|%%END%%)', each,re.MULTILINE)
    startDate = re.search(r'^(%%START_DATE%%:)((.*\s)*?)(%%SALARY%%:|%%QUALIFICATIONS%%:|%%JOB_PROCEDURE%%:|%%ABOUT_COMPANY%%:|%%DEADLINE%%:|%%JOB_DESCRIPTION%%:|%%LOCATION%%:|%%JOB_RESPONSIBILITIES%%:|%%JOB_TITLE%%:|OPEN TO/|%%END%%)', each,re.MULTILINE)
    
    # Contains all individual applications. This dictionary is appended to the list "listing"
    resumes = {}
    
    ###### JOB ID ######    
    output_xml.write("\t<listing id='"+id+"'>\n")
    resumes["_id"] = id
    
    ###### JOB TITLE ######
    if jobTitle == None:
        output_xml.write("\t\t<title>N/A</title>\n")
        resumes["title"] = "N/A"
    else:
        title = jobTitle.group(2)
        if title.isspace():
            output_xml.write("\t\t<title>N/A</title>\n")
            resumes["title"] = "N/A"
        else:
            title = title.replace("\n"," ")
            title = title.strip()
            output_xml.write("\t\t<title>"+ title+ "</title>\n")
            resumes["title"] = title
            
    ###### LOCATION ######
    if location == None:
        output_xml.write("\t\t<location>N/A</location>\n")
        resumes["location"] = "N/A"
    else:
        loc = location.group(2)
        if loc.isspace():
            output_xml.write("\t\t<location>N/A</location>\n")
            resumes["location"] = "N/A"
        else:
            loc = loc.replace("\n"," ")
            loc = loc.strip()
            output_xml.write("\t\t<location>"+ loc+ "</location>\n")
            resumes["location"] = loc
            
    ###### JOB DESCRIPTION ######
    if jobDescription == None:
        output_xml.write("\t\t<job_descriptions>N/A</job_descriptions>\n")
        resumes["job_descriptions"] = "N/A"
    else:
        description = jobDescription.group(2)
        if description.isspace():
            output_xml.write("\t\t<job_descriptions>N/A</job_descriptions>\n")
            resumes["job_descriptions"] = "N/A"
        else:
            description = description.replace("\n"," ")
            description = description.strip()
            output_xml.write("\t\t<job_descriptions>"+ description+ "</job_descriptions>\n")
            resumes["job_descriptions"] = description
            
    ###### JOB RESPONSIBILITIES ######
    if jobResponsibility == None:
        output_xml.write("\t\t<job_responsibilities>N/A</job_responsibilities>\n")
        resumes["job_responsibilities"] = "N/A"
    else:
        responsibility = jobResponsibility.group(2)
        job_responsibilities = {}
        if responsibility.isspace():
            output_xml.write("\t\t<job_responsibilities>N/A</job_responsibilities>\n")
            resumes["job_responsibilities"] = "N/A"
        else:
            responsibility = responsibility.strip()
            responsibility = "\n"+responsibility
            responsibility = responsibility.split("\n-")
            if not responsibility[0]:
                responsibility.pop(0)
            output_xml.write("\t\t<job_responsibilities>\n")
            job_responsibilities["responsibility"] = responsibility
            resumes["job_responsibilities"] = job_responsibilities
            for each in responsibility:
                each = each.strip()
                each = each.replace("\n"," ")
                output_xml.write("\t\t\t<job_responsibility>"+each+"</job_responsibility>\n")
            output_xml.write("\t\t</job_responsibilities>\n")
            
    ###### QUALIFICATIONS ######
    if jobQualifications == None:
        output_xml.write("\t\t<required_qualifications>N/A</required_qualifications>\n")
        resumes["required_qualifications"] = "N/A"
    else:
        qualifications = jobQualifications.group(2)
        required_qualifications = {}
        if qualifications.isspace():
            output_xml.write("\t\t<required_qualifications>N/A</required_qualifications>\n")
            resumes["required_qualifications"] = "N/A"
        else:
            qualifications = qualifications.strip()
            qualifications = "\n"+qualifications
            qualifications = qualifications.split("\n-")
            if not qualifications[0]:
                qualifications.pop(0)
            output_xml.write("\t\t<required_qualifications>\n")
            required_qualifications["qualifications"] = qualifications
            resumes["required_qualifications"] = required_qualifications
            for each in qualifications:
                each = each.strip()
                each = each.replace("\n"," ")
                output_xml.write("\t\t\t<qualification>"+each+"</qualification>\n")
            output_xml.write("\t\t</required_qualifications>\n")
            
    ###### SALARY ######
    if salary == None:
        output_xml.write("\t\t<salary>N/A</salary>\n")
        resumes["salary"] = "N/A"
    else:
        sal = salary.group(2)
        if sal.isspace():
            output_xml.write("\t\t<salary>N/A</salary>\n")
            resumes["salary"] = "N/A"
        else:
            sal = sal.replace("\n"," ")
            sal = sal.strip()
            #sal = sal.replace("%%SALARY%%","salary")
            output_xml.write("\t\t<salary>"+ sal+ "</salary>\n")
            resumes["salary"] = sal
    
    
    ###### JOB PROCEDURE ######
    if jobProcedure == None:
        output_xml.write("\t\t<application_procedure>N/A</application_procedure>\n")
        resumes["application_procedure"] = "N/A"
    else:
        procedure = jobProcedure.group(2)
        if procedure.isspace():
            output_xml.write("\t\t<application_procedure>N/A</application_procedure>\n")
            resumes["application_procedure"] = "N/A"
        else:
            procedure = procedure.replace("\n"," ")
            procedure = procedure.strip()
            output_xml.write("\t\t<application_procedure>"+ procedure+ "</application_procedure>\n")
            resumes["application_procedure"] = procedure
            
    
     ###### START DATE ######
    if startDate == None:
        output_xml.write("\t\t<start_date>N/A</start_date>\n")
        resumes["start_date"] = "N/A"
    else:
        start = startDate.group(2)
        if start.isspace():
            output_xml.write("\t\t<start_date>N/A</start_date>\n")
            resumes["start_date"] = "N/A"
        else:
            start = start.replace("\n"," ")
            start = start.strip()
            output_xml.write("\t\t<start_date>"+ start+ "</start_date>\n")
            resumes["start_date"] = start
            
            
    ###### DEADLINE ######
    if deadLine == None:
        output_xml.write("\t\t<application_deadline>N/A</application_deadline>\n")
        resumes["application_deadline"] = "N/A"
    else:
        deadline = deadLine.group(2)
        if deadline.isspace():
            output_xml.write("\t\t<application_deadline>N/A</application_deadline>\n")
            resumes["application_deadline"] = "N/A"
        else:
            deadline = deadline.replace("\n"," ")
            deadline = deadline.strip()
            output_xml.write("\t\t<application_deadline>"+ deadline+ "</application_deadline>\n")
            resumes["application_deadline"] = deadline
            
            
    ###### ABOUT COMPANY ######
    if aboutCompany == None:
        output_xml.write("\t\t<about_company>N/A</about_company>\n")
        resumes["about_company"] = "N/A"
    else:
        about = aboutCompany.group(2)
        if about.isspace():
            output_xml.write("\t\t<about_company>N/A</about_company>\n")
            resumes["about_company"] = "N/A"
        else:
            about = about.replace("\n"," ")
            about = about.strip()
            output_xml.write("\t\t<about_company>"+ about+ "</about_company>\n")
            resumes["about_company"] = about
    
   
    output_xml.write("\t</listing>\n")
    listing.append(resumes)
    
output_xml.write("</listings>\n")

listings["listing"] = listing
listingJSON["listings"] = listings

# Create JSON output
outputJSON = json.dumps(listingJSON)

output_json.write(outputJSON)

output_xml.close()
output_json.close()
print("OK")
```
